```r
library("ggplot2")
```

```r
df <- read.csv("perf_vs_cplx_data")

df$Form = factor(df$Form, levels=c('General Form', 'CNF','d-DNNF')) 
df$cplx = factor(df$cplx, levels=c('Low','Moderate','High'))
# colors <- c("#003671", "#FFAF00",  "#BFF8FB", "#B61A1C") 
colors <- c("#003671", "#FFAF00",  "#74c476", "#B61A1C")


g <- ggplot(df, aes(x=cplx, y=mean, fill=Form)) + geom_bar(stat="identity", position=position_dodge(),colour='black') + geom_errorbar(aes(ymin=mean-se, ymax=mean+se), width=.3, position=position_dodge(.9)) + 
  theme_bw() + ylab("Test Performanc Acc") + xlab("Complexity")+ theme(legend.position="top",
#                      legend.title = element_text(size=16, face="bold"),
                     legend.title = element_blank(),
                     legend.spacing.x = unit(0.5, 'cm'),
                     strip.background = element_rect(fill="#FFFFFF"), 
                     strip.text = element_text(size=15), 
                     axis.text = element_text(size=15),
                     #axis.title.x = element_text(size=14, vjust=-0.2),
#                      axis.title.x = element_blank(),
                     axis.title.y = element_text(size=15, vjust=0.8),
                     legend.text = element_text(size=15),
                     title = element_text(size=18, vjust=1),
                     panel.grid = element_blank()) + scale_fill_manual(values = colors) + coord_cartesian(ylim = c(0.6, 1)) 
g

ggsave('perf_vs_cplx.pdf',width = 6, height = 4.5)





```

```r

#[0.33897823 4.90414974 0.58007464]
f <- function(x) 0.33897823 * exp(-4.90414974 * x) + 0.58007464
df <- read.csv("sat_vs_dist_data")
df$form = factor(df$form, levels=c('General Form', 'CNF','d-DNNF')) 
g <- ggplot(df, aes(x=dist, y=sat, shape=form, color=form)) + geom_point(size=4) + stat_function(fun=f, colour=colors[3],size=1)+scale_color_manual(values=colors)+theme_bw() + labs(x="Squared Euclidean Distance (normalized)",y="Satisfiability (percentage)")+ theme(legend.position="top",
#                      legend.title = element_text(size=16, face="bold"),
                     legend.title = element_blank(),
                     legend.spacing.x = unit(0.5, 'cm'),
                     strip.background = element_rect(fill="#FFFFFF"), 
                     strip.text = element_text(size=15), 
                     axis.text = element_text(size=15),
                     #axis.title.x = element_text(size=14, vjust=-0.2),
#                      axis.title.x = element_blank(),
                     axis.title.y = element_text(size=15, vjust=0.8),
                     legend.text = element_text(size=15),
                     title = element_text(size=18, vjust=1),
                     panel.grid = element_blank()) + scale_fill_manual(values = colors)# + coord_cartesian(ylim = c(0.6, 1)) 
g
ggsave('sat_vs_dist.pdf',width = 6, height = 4.5)


```

```r
library("plyr")
df <- read.csv("perf_vs_iter_data")
df$form = factor(df$form, levels=c('General Form', 'CNF','d-DNNF','d-DNNF w/ HE & SR')) 
df$color <- mapvalues(df$form, from=c('General Form', 'CNF','d-DNNF','d-DNNF w/ HE & SR'), to=colors)

g <- ggplot(df,aes(x=iter, y=mean, group=form,colour=form))+geom_line(size=1) +theme_bw() +geom_ribbon(aes(ymin=df$mean-df$se, ymax=df$mean+df$se),alpha=0.3,group = df$form,fill = df$color,
              colour=NA)+labs(x="Iteration",y="Train Prediction Loss")+theme(legend.position="top",
#                      legend.title = element_text(size=16, face="bold"),
                     legend.title = element_blank(),
                     legend.spacing.x = unit(0.1, 'cm'),
                     strip.background = element_rect(fill="#FFFFFF"), 
                     strip.text = element_text(size=15), 
                     axis.text = element_text(size=15),
                     #axis.title.x = element_text(size=14, vjust=-0.2),
#                      axis.title.x = element_blank(),
                     axis.title.y = element_text(size=15, vjust=0.8),
                     legend.text = element_text(size=12),
                     title = element_text(size=18, vjust=1),
                     panel.grid = element_blank()) + scale_color_manual(values = colors)+scale_fill_manual(values = colors)
g

ggsave('perf_vs_iter.pdf')#,width = 6, height = 4.5)

```
