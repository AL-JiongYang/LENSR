```python
avg_window =10000
# dataset = ["_0303","_0603","_1206","_1209","_0906","_0909","_0912","_0306","_0309","_0312","_0606","_0612","_0609"]#,"_new"]
dataset = ["_0303","_0306","_0309","_0312","_0603","_0606","_0609","_0612","_0906","_0909","_0912","_1206","_1209"]
print (len(dataset))
# names = ['general', 'cnf', 'cnf_ind','ddnnf','ddnnf_reg','ddnnf_ind','ddnnf_reg_ind']
names = ['general', 'cnf', 'ddnnf','ddnnf_reg_ind']
form_name = ['General Form','CNF','d-DNNF','d-DNNF w/ HE & SR']
seed_names = ['.seed1','.seed2']

for d in dataset:
#     if d == '_0303':
#     if d != '_1209' and d != '_0909' and d!='_0912' and d != '_0312':
    if True:
        perf_vs_iter_data = {'form':[],'mean':[],'se':[],'iter':[]}
        plt.clf()
        for i in range(len(names)):
            n = names[i]
            if 'general' in n:
                n_name = 'general'
                reg_name = 'reg0.1'
                ind_name = ''
            elif 'cnf' in n:
                n_name = 'cnf'
                reg_name = 'reg0.0'
                ind_name = 'ind.' if 'ind' in n else ''
            elif 'ddnnf' in n:
                n_name = 'ddnnf'
                reg_name = 'reg0.1' if 'reg' in n else 'reg0.0'
                ind_name = 'ind.' if 'ind' in n else ''

            avg_all = []
            min_length = None
            for seed_name in seed_names:
                loss_by_iter = pk.load(open(f'{n_name}{d}.{reg_name}.{ind_name}cls0.1{seed_name}.loss_by_iter','rb'))

                avg = np.convolve(loss_by_iter, np.ones((avg_window,))/avg_window, mode='valid')
                avg_all.append(avg)
                if min_length is None:
                    min_length = len(avg)
                else:
                    if min_length > len(avg):
                        min_length = len(avg)
            for avg in range(len(avg_all)):
                avg_all[avg] = avg_all[avg][:min_length]

            mean = np.mean(avg_all,axis=0)
    #         err = np.std(avg,axis=0)
            err = scipy.stats.sem(avg_all,axis=0)
        
            for item in range(len(mean)):
                perf_vs_iter_data['mean'].append(mean[item])
                perf_vs_iter_data['se'].append(err[item])
                perf_vs_iter_data['form'].append(form_name[i])
                perf_vs_iter_data['iter'].append(item)
            
            plt.plot(list(range(0,len(mean))), mean, label=labels[i],color=color[i])
            plt.fill_between(list(range(0,len(mean))),mean-err,mean+err,facecolor=color[i],alpha=0.5)
        plt.title('Train Prediction Loss v.s. Iteration - Dataset '+d[1:])
        plt.xlabel('Iteration')
        plt.ylabel('Train Prediction Loss')
        plt.legend()
        if d == '_0612':
            plt.savefig("perf_vs_iter.pdf")
        plt.show()

perf_vs_iter_data = pd.DataFrame(perf_vs_iter_data)
perf_vs_iter_data.to_csv('perf_vs_iter_data', index=False)
```

```python
from MulticoreTSNE import MulticoreTSNE as TSNE
from sklearn.decomposition import PCA

color = ['r','b','g','y']
dataset = '_0606'
# names = ['general', 'cnf', 'cnf_ind','ddnnf','ddnnf_reg','ddnnf_ind','ddnnf_reg_ind']
names = ['general', 'cnf', 'ddnnf']
# names = ['general']

seed_name = '.seed1'

tsne_embedding = []
sol_id_all = []
for n in names:
    sol_id_all.append([])
    n_name = n
    if 'general' in n:
        reg_name = 'reg0.1'
        ind_name = ''
    elif 'cnf' in n:
        n_name = 'cnf'
        reg_name = 'reg0.0'
        ind_name = 'ind.' if 'ind' in n else ''
    elif 'ddnnf' in n:
        n_name = 'ddnnf'
        reg_name = 'reg0.1' if 'reg' in n else 'reg0.0'
        ind_name = 'ind.' if 'ind' in n else ''
    quality_data = pk.load(open(f'../quality_plots/{n_name}{dataset}.{reg_name}.{ind_name}cls0.1{seed_name}.sat_bin.quality.pk','rb'))
    fml_sol = np.array(quality_data['fml_emb']+quality_data['sol_emb'])
    
    for fml_id in range(len(quality_data['fml_emb'])):
        sat_sol_id = []
        unsat_sol_id = []
        for b in range(len(quality_data['sol_emb'])):
            if quality_data['sat'][b][fml_id] >= 1:
                sat_sol_id.append(b)
            else:
                unsat_sol_id.append(b)
        
        sol_id_all[-1].append([sat_sol_id,unsat_sol_id])
    
    print (fml_sol.shape)
    pca = PCA(n_components=100)
    fml_sol = pca.fit_transform(fml_sol)
    print (pca.explained_variance_ratio_)
    print (fml_sol.shape)

    tsne = TSNE(n_jobs=4,learning_rate=800,n_components=2,perplexity=30)
    Y = tsne.fit_transform(fml_sol)
    tsne_embedding.append(Y)
    
    
```
